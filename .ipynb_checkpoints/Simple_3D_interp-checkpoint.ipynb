```python
import numpy as np
from scipy.interpolate import LinearNDInterpolator
import matplotlib.pyplot as plt
```

## I - general implementation of LinearNDInterpolator for interpolation of channels to 3D matrix
### 1) create locations (2D for simplicity) and signals for the channels

```python
n_channels = 128
T = 30
channels_pos = np.random.rand(n_channels,2)
T_range = np.arange(T)
channels = np.random.rand(n_channels, T)
```

### 2) create the variables necessary for LinearNDInterpolator and define the grid (x,y) on which you want to interpolate in output

```python
# 3D points (x,y,t)
points = np.array([[channels_pos[i,0],channels_pos[i,1],t] for t in T_range for i in range(n_channels)])
# the value of the EEG measurement at these (x,y,t)
values = np.array([channels[i,t] for t in T_range for i in range(n_channels)])

# output grid: X*Y*T_range
X = np.linspace(0,1,15)
Y = np.linspace(0,1,15)
# flatten this 3D grid in list of size-3 vectors (x,y,t)
request = np.array([[x,y,t] for t in T_range for y in Y for x in X])
# save the indicies for later
indices = np.array([[i_x, i_y] for t in T_range for i_y in range(len(Y)) for i_x in range(len(X))])
```

### 3) Get the interpolation values on the "request" 3D points
The result is a list of values that will need to be reshaped in a 3D matrix (done later)

```python
# create the interpolator
linInter = LinearNDInterpolator(points, values, fill_value=0)
# interpolate
result = linInter(request)
```

### 4) Understand where interpolation is performed
LinearNDInterpolator enables to interpolate inside a convex set containing all the samples, and the outside value needs to be assigned the 0 value (the default value is NaN as you can see below)

```python
# observe the convex set on which interpolation is limited

# plot the channels locations
plt.scatter(channels_pos[:,0],channels_pos[:,1],color='b')

# plot the locations of the grid (x,y,:) at which interpolation was possible (green) or not (red)
nonan_indices = np.unique(indices[result>0], axis=0)
nan_indices = np.unique(indices[result==0], axis=0)
plt.scatter(X[nonan_indices[:,0]],Y[nonan_indices[:,1]], color='g')
plt.scatter(X[nan_indices[:,0]],Y[nan_indices[:,1]], color='r')
```

### 5) Understand how interpolation is performed
The algorithm uses the Quickhull algorithm to find the convex hull of the samples (see image below) and then is given with linear barycentric interpolation on triangles. (source: Scipy)

```python
help(LinearNDInterpolator)
```

```python
def f(t):
    return 1/(1+5*t**2)
```

```python
# create a simpler dataset with 5 channels
n_channels = 5
T = 30
channels_pos = np.array([[0,0],[0,0.5],[0,1],[1,0],[1,1]])
T_range = np.arange(T)
channels = np.array([f(0.5*T_range),
                     f(0.5*T_range-4),
                    f(0.5*T_range-8),
                    0*f(T_range),
                    0*f(T_range)])
plt.plot(channels.T);
```

```python
points = np.array([[channels_pos[i,0],channels_pos[i,1],t] for t in T_range for i in range(n_channels)])
values = np.array([channels[i,t] for t in T_range for i in range(n_channels)])
nX = 5
X = np.linspace(0,1,nX)
nY = 10
Y = np.linspace(0,1,nY)
request = np.array([[x,y,t] for t in T_range for y in Y for x in X])
indices = np.array([[i_x, i_y] for t in T_range for i_y in range(nY) for i_x in range(nX)])
```

```python
# create the interpolator
linInter = LinearNDInterpolator(points, values, fill_value=0)
# interpolate
result = linInter(request)
```

#### How to reshape the output

```python
# reshape the output to a 3D matrix such that
# reshaped[i,j,k] is the value of interpolation at time k and at location (X[i], Y[j])
reshaped = result.reshape((T,nY,nX)).transpose((2,1,0))
```

```python
plt.figure(figsize=(10,8))
for j in range(nY):
    plt.subplot(4,3,j+1)
    plt.plot(channels.T,'--')
    plt.plot(reshaped[0,j,:])
```

### 6) How should we include time as the 3rd dimension? 
In the code above, the time dimension is treated as a 3rd spatial dimension to do a linear interpolation. This enables to reproduce traveling waves (for example) and induce a delay between nearby interpolation points. 
To modulate this effect, the time dimension can be expanded or reduced by mutliplying T_range by a factor Alpha when the variables "points" and "request" are defined.
- To maximize the effect of the time in this interpolation, Alpha should be small.
- To eliminate the time in the 3D linear interpolation, Alpha should be large.

#### 6.1: maximize the effect of time

```python
Alpha = 0.01
Beta = 10

# create a simpler dataset with 5 channels
n_channels = 5
T = 30
channels_pos = np.array([[0,0],[0,0.5],[0,1],[1,0],[1,1]])
T_range = np.arange(T)
channels = np.array([f(0.5*T_range),
                     f(0.5*T_range-4),
                    f(0.5*T_range-8),
                    0*f(T_range),
                    0*f(T_range)])

############# Line below: first use of Alpha
points = np.array([[channels_pos[i,0],channels_pos[i,1],t+Alpha*(2*np.random.rand()-1)] for t in Alpha*T_range for i in range(n_channels)])
values = np.array([channels[i,i_t] for i_t in range(T) for i in range(n_channels)])
nX = 5
X = np.linspace(0,1,nX)
nY = 10
Y = np.linspace(0,1,nY)
############# Line below: second use of Alpha
request = np.array([[x,y,t] for t in Alpha*T_range for y in Y for x in X])
indices = np.array([[i_x, i_y] for t in T_range for i_y in range(nY) for i_x in range(nX)])

# create the interpolator
linInter = LinearNDInterpolator(points, values, fill_value=0, rescale=False)
# interpolate
result = linInter(request)

# reshape the output to a 3D matrix such that
# reshaped[i,j,k] is the value of interpolation at time k and at location (X[i], Y[j])
reshaped = result.reshape((T,nY,nX)).transpose((2,1,0))

plt.figure(figsize=(10,8))
for j in range(nY):
    plt.subplot(4,3,j+1)
    plt.plot(channels.T,'--')
    plt.plot(reshaped[0,j,:])
```
